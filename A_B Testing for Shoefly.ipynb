```python
import codecademylib
import pandas as pd

ad_clicks = pd.read_csv('ad_clicks.csv')
print(ad_clicks.head())

views_per_source = ad_clicks.groupby("utm_source").user_id.count().reset_index()
print(views_per_source)

ad_clicks["is_click"] = ad_clicks.ad_click_timestamp.notnull()
print(ad_clicks.head())

clicks_by_source = ad_clicks.groupby(["utm_source", "is_click"]).user_id.count().reset_index()
print(clicks_by_source)

clicks_pivot = clicks_by_source.pivot(
  columns = "is_click",
  index = "utm_source",
  values = "user_id"
).reset_index()
print(clicks_pivot)

clicks_pivot["percent_clicked"] = clicks_pivot[True] / (clicks_pivot[True] + clicks_pivot[False])
print(clicks_pivot)



a_b_count = ad_clicks.groupby("experimental_group").user_id.count().reset_index()
print (a_b_count)

a_b_is_clicked = ad_clicks.groupby(["experimental_group", "is_click"]).user_id.count().reset_index()
print(a_b_is_clicked)

a_b_clicked_pivot = a_b_is_clicked.pivot(
  columns = "is_click",
  index = "experimental_group",
  values = "user_id"
).reset_index()
print(a_b_clicked_pivot)

a_clicks = ad_clicks[ad_clicks.experimental_group == "A"]
b_clicks = ad_clicks[ad_clicks.experimental_group == "B"]
print(a_clicks)
print(b_clicks)


a_by_day_click = a_clicks.groupby(["day", "is_click"]).user_id.count().reset_index()
b_by_day_click = b_clicks.groupby(["day", "is_click"]).user_id.count().reset_index()
print(a_by_day_click)
print(b_by_day_click)
a_by_day_click_pivot = a_by_day_click.pivot(
  columns = "is_click",
  index = "day",
  values = "user_id"
).reset_index()
# print(a_by_day_click_pivot)

a_by_day_click_pivot["percent_clicked"] = a_by_day_click_pivot[True] / (a_by_day_click_pivot[True] + a_by_day_click_pivot[False])
print(a_by_day_click_pivot)

b_by_day_click_pivot = b_by_day_click.pivot(
  columns = "is_click",
  index = "day",
  values = "user_id"
).reset_index()
# print(b_by_day_click_pivot)
b_by_day_click_pivot["percent_clicked"] = b_by_day_click_pivot[True] / (b_by_day_click_pivot[True] + b_by_day_click_pivot[False])
print(b_by_day_click_pivot)
```
